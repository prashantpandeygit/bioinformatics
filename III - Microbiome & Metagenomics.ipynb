**Microbiome**

A microbiome is the collection of microorganisms that live in a specific environment. (eg: gut microbiome, skin microbiome, etc.)

Each microbiome contains: 

1. Bacteria
2. Viruses
3. Fungi
4. Protozoa
5. Archaea

**Metagenomics**

Metagenomics is the study of the microbiome, or more like sequencing DNA from a microbial community.

Sample -> DNA extraction -> DNA sequencing -> Classification -> Taxanomic Profile


**Taxanomic Profile Example**


\begin{cases}
\textit{Escherichia coli} & 25\% \\
\textit{Bacteroides fragilis} & 40\% \\
\textit{Streptococcus} & 10\% \\
\text{Other} & 25\%
\end{cases}

**Taxonomy Heirarchy**

Kingdom > Phylum > Class > Order > Family > Genus > Species


**Example:**
\begin{aligned}
\textbf{Domain:} & \ \text{Bacteria} \\
\textbf{Phylum:} & \ \textit{Proteobacteria} \\
\textbf{Genus:} & \ \textit{Escherichia} \\
\textbf{Species:} & \ \textit{Escherichia coli}
\end{aligned}

**For Datasets, taxonomic is represented as:**


\begin{aligned}
\textbf{Genus\_A} &: 0.32 \\
\textbf{Genus\_B} &: 0.12 \\
\textbf{Genus\_C} &: 0.04 \\
& \vdots
\end{aligned}
